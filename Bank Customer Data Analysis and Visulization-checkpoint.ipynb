```python
import seaborn as sns

fig, axes = plt.subplots(3, 3, figsize=(15, 15))
# Age Distribution (Histogram)
sns.histplot(df['age'], kde=True, ax=axes[0, 0])
axes[0, 0].set_title('Age Distribution')

# Job Distribution (Bar Chart)
sns.countplot(data=df, x='job', ax=axes[0, 1])
axes[0, 1].set_title('Job Distribution')
axes[0, 1].tick_params(axis='x', rotation=45)

# Marital Status Distribution (Bar Chart)
sns.countplot(data=df, x='marital', ax=axes[0, 2])
axes[0, 2].set_title('Marital Status Distribution')

# Education Distribution (Bar Chart)
sns.countplot(data=df, x='education', ax=axes[1, 0])
axes[1, 0].set_title('Education Distribution')
axes[1, 0].tick_params(axis='x', rotation=45)

# Default Distribution (Bar Chart)
sns.countplot(data=df, x='default', ax=axes[1, 1])
axes[1, 1].set_title('Default Distribution')

# Balance Distribution (Histogram)
sns.histplot(df['balance'], kde=True, ax=axes[1, 2])
axes[1, 2].set_title('Balance Distribution')

# Housing Loan Distribution (Bar Chart)
sns.countplot(data=df, x='housing', ax=axes[2, 0])
axes[2, 0].set_title('Housing Loan Distribution')

# Personal Loan Distribution (Bar Chart)
sns.countplot(data=df, x='loan', ax=axes[2, 1])
axes[2, 1].set_title('Personal Loan Distribution')

# Contact Method Distribution (Bar Chart)
sns.countplot(data=df, x='contact', ax=axes[2, 2])
axes[2, 2].set_title('Contact Method Distribution')
axes[2, 2].tick_params(axis='x', rotation=45)
plt.tight_layout()
plt.show()

```

```python

fig, ax = plt.subplots(figsize=(10, 6))
sns.scatterplot(data=df, x='age', y='balance', ax=ax, alpha=0.5)
sns.regplot(data=df, x='age', y='balance', ax=ax, scatter=False, color='red')
start_age = 10
interval_width = 5
max_age = df['age'].max()
age_intervals = np.arange(start_age, max_age + interval_width, interval_width)

# Set x-axis labels to age intervals
ax.set_xticks(age_intervals)
ax.set_xticklabels([f"{age}-{age+interval_width}" for age in age_intervals])
ax.set_title('Age vs. Balance Scatter Plot')
ax.set_xlabel('Age Intervals')
ax.set_ylabel('Balance')
ax.legend(['Regression Line'])
sns.set(style="whitegrid") 
plt.tight_layout()
plt.show()

```

```python

from scipy.stats import ttest_ind
data = {
    'age': np.random.randint(20, 60, 1000),
    'job': np.random.choice(['Admin', 'Technician', 'Manager', 'Student', 'Retired'], 1000),
    'marital': np.random.choice(['Married', 'Single', 'Divorced'], 1000),
    'education': np.random.choice(['Primary', 'Secondary', 'Tertiary'], 1000),
    'default': np.random.choice(['Yes', 'No'], 1000),
    'balance': np.random.randint(-500, 5000, 1000),
    'housing': np.random.choice(['Yes', 'No'], 1000),
    'loan': np.random.choice(['Yes', 'No'], 1000),
    'contact': np.random.choice(['Cellular', 'Telephone', 'Unknown'], 1000),
    'day': np.random.randint(1, 31, 1000),
    'month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 1000),
    'duration': np.random.randint(10, 1000, 1000),
    'campaign': np.random.randint(1, 10, 1000),
    'OperatingSystems': np.random.randint(1, 5, 1000),
    'Browser': np.random.randint(1, 4, 1000),
    'Region': np.random.choice(['North', 'South', 'East', 'West'], 1000),
    'TrafficType': np.random.randint(1, 10, 1000),
    'Weekend': np.random.choice(['Yes', 'No'], 1000)
}

df = pd.DataFrame(data)
summary_stats = df.describe()
plt.figure(figsize=(12, 6))
sns.histplot(df['age'], kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
correlation_matrix = df.select_dtypes(include='number').corr()
print(correlation_matrix)
#  Frequency Distributions
job_distribution = df['job'].value_counts()
plt.figure(figsize=(12, 6))
sns.barplot(x=job_distribution.index, y=job_distribution.values)
plt.title('Job Distribution')
plt.xlabel('Job')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

# Grouped Analysis
education_grouped = df.groupby('education')['balance'].mean()
print(education_grouped)
monthly_balance = df.groupby('month')['balance'].mean()
plt.figure(figsize=(10, 5))
sns.lineplot(data=df, x='month', y='balance')
plt.title('Monthly Balance Trend')
plt.xlabel('Month')
plt.ylabel('Average Balance')
plt.show()
group1 = df[df['education'] == 'Primary']['balance']
group2 = df[df['education'] == 'Secondary']['balance']

t_stat, p_value = ttest_ind(group1, group2)
print(f"t-statistic: {t_stat}")
print(f"P-value: {p_value}")


```

```python
df['contact'] = df['contact'].replace(['Unknown'], 'unknown')
unknown_contacts_percentage = (df['contact'].value_counts(normalize=True).get('unknown', 0) * 100)

# Create a DataFrame for the pie chart
contact_data = pd.DataFrame({'Contact Type': ['Unknown', 'Known'],
                             'Percentage': [unknown_contacts_percentage, 100 - unknown_contacts_percentage]})
plt.figure(figsize=(6, 6))
plt.pie(contact_data['Percentage'], labels=contact_data['Contact Type'], autopct='%1.1f%%', startangle=90)
plt.title('Percentage of Unknown Contacts')
plt.show()
```

```python

zero_or_negative_balance_with_loan = df[(df['balance'] <= 0) & (df['loan'] == 'Yes')]
count_with_zero_or_negative_balance_and_loan = len(zero_or_negative_balance_with_loan)
count_all_with_zero_or_negative_balance = len(df[df['balance'] <= 0])

# Calculate the count of account holders with positive balance
count_all_with_positive_balance = len(df[df['balance'] > 0])
data = pd.DataFrame({
    'Category': ['Zero/Negative Balance with Loan', 'Zero/Negative Balance without Loan', 'Positive Balance'],
    'Count': [count_with_zero_or_negative_balance_and_loan, count_all_with_zero_or_negative_balance, count_all_with_positive_balance]
})

# Plotting a pie chart to show the distribution
plt.figure(figsize=(6, 6))
plt.pie(data['Count'], labels=data['Category'], autopct='%1.1f%%', startangle=90)
plt.title('Account Holders with Zero/Negative Balance and Loan')
plt.show()
```
